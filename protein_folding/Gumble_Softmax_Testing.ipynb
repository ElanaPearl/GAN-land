This plots the distributions generated by the generator. You can see that over time it creates 'one-hot'-esque predictions.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
for i in range(9):
    plt.hist(pd.read_csv('out/00{}.txt'.format(i), header=None).T, alpha=0.5)
    plt.title(i)
    plt.yscale('log')
    plt.show()
```
